#### Fix remaining columns to integers or floats as needed

```python
# use a helper function to clean and convert multiple columns

def clean_and_convert(df, cols, dtype="float"):
    """
    Cleans columns (removes text placeholders like 'nan' or 'Missing value')
    and converts them to numeric dtype ('float' or 'Int64').
    """
    for col in cols:
        if col not in df.columns:
            print(f" Column '{col}' not found, skipping.")
            continue

        # Clean placeholders
        df[col] = (
            df[col]
            .astype(str)
            .str.strip()
            .replace(r'^(?:nan|na|none|null|missing value|\s*)$', np.nan, regex=True)
        )

        # Convert to numeric
        df[col] = pd.to_numeric(df[col], errors='coerce')

        # Convert dtype
        if dtype.lower() == "int":
            df[col] = df[col].round().astype("Int64")
        else:
            df[col] = df[col].astype(float)

        # Optional check
        print(f"{col}: {df[col].dtype}, non-null={df[col].notna().sum()}")
```

```python
# Example groups
int_cols = ['Parity', 'Gravida', 'Delivery_Month', 'Gest_Age_Birth', 'Gest_Age_Baseline',
            'Estimated Bloodloss', 'baby_birthweight', 'd5weightchange'] # could be floats but since continuous values not expected, convert to int
float_cols = [
    'maternal_height', 'maternal_weight', 'prepreg BMI', 
    'baselinetimedec', 'gestagebirth_weeks', 
    'percentd5weightchange'
]

# Convert them all
clean_and_convert(meta_shared, int_cols, dtype="int")
clean_and_convert(meta_shared, float_cols, dtype="float")
```
